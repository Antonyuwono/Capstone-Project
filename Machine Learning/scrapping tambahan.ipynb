```python
import requests
from bs4 import BeautifulSoup
import os

save_dir = "images/"
if not os.path.exists(save_dir):
    os.makedirs(save_dir)

query = "benteng vredeburg"
num_images_to_download = 100
downloaded_images = 0
page = 0

while downloaded_images < num_images_to_download:
    search_url = f"https://www.google.com/search?q={query}&tbm=isch&start={page * 20}"
    response = requests.get(search_url)
    soup = BeautifulSoup(response.content, "html.parser")
    image_tags = soup.find_all("img")

    if not image_tags:
        print("No more images found.")
        break

    for img_tag in image_tags:
        if downloaded_images >= num_images_to_download:
            break
        try:
            image_url = img_tag["src"]
            image_data = requests.get(image_url).content
            with open(os.path.join(save_dir, f"{query}_{downloaded_images}.jpg"), "wb") as f:
                f.write(image_data)
            downloaded_images += 1
        except Exception as e:
            print(f"Could not download image {img_tag}: {e}")

    page += 1

print(f"Downloaded {downloaded_images} images.")

# Zip the folder
import zipfile

zipf = zipfile.ZipFile('images.zip', 'w', zipfile.ZIP_DEFLATED)
for root, dirs, files in os.walk(save_dir):
    for file in files:
        zipf.write(os.path.join(root, file))
zipf.close()

print("Images have been zipped.")

# Download the zipped folder (only applicable in environments like Google Colab)
try:
    from google.colab import files
    files.download('images.zip')
except ImportError:
    print("Google Colab not detected. Skipping download step.")
```
