# **Convert to TensorflowLite**

```python
import tensorflow as tf

# Load the Keras model without compilation
model = tf.keras.models.load_model('/content/drive/MyDrive/Historia/building_recognition_model.h5', compile=False)
```

```python
# Convert the model to TensorFlow Lite format
converter = tf.lite.TFLiteConverter.from_keras_model(model)
tflite_model = converter.convert()

# Save the TensorFlow Lite model
with open('/content/drive/My Drive/building_recognition_model.tflite', 'wb') as f:
    f.write(tflite_model)
```
